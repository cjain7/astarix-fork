```python
def plot_time_by_read_distr(alignments_tsv, png):
    df = pd.read_csv(alignments_tsv, sep='\t')
    #df.head()
    fig, ax = plt.subplots()
    s = df.pushed / df.pushed.sum()
    s.cumsum().plot(ax=ax)
    ax.set_xlabel('reads', size=18)
    ax.set_ylabel('time part', size=18)
    fig.savefig(png, dpi=300, bbox_inches='tight')

f = '/home/pesho/work/astarix-gitlab/tmp/ecoli_head1000000_linear_long_reads/astar-waymark_errors_2/alignments.tsv'
plot_time_by_read_distr(f, 'distr.png')
```
