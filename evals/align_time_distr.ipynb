```python
from header import *
plt.rcParams.update({'font.size': 14})
```

```python
def plot_time_by_read_distr(alignments_tsv, png):
    df = pd.read_csv(alignments_tsv, sep='\t')
    #df.head()
    fig, (ax1, ax2) = plt.subplots(1,2)
    s = df.pushed / df.pushed.sum()
    s.sort_values(ignore_index=True, inplace=True)
    
    s.cumsum().plot(ax=ax1)
    ax1.set_xlabel('#reads', size=18)
    ax1.set_ylabel('time part', size=18)
    
    s.plot.hist(ax=ax2, bins=20)
    ax2.set_xlabel('time', size=18)
    ax2.set_ylabel('#reads', size=18)
    
    fig.savefig(png, dpi=300, bbox_inches='tight')
```
