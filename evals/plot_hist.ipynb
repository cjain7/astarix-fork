```python
from header import *
plt.rcParams.update({'font.size': 14})
```

```python
def draw_hist(hist_fn, png, MAX_I=15, ylog=True):
    df = pd.read_csv(hist_fn, sep='\t')
    fig, ax = plt.subplots()
    for index, row in df.iterrows():
        ax.plot(df.columns[1:MAX_I], row[1:MAX_I], c='g', alpha=0.05)
    ax.plot(df.columns[1:MAX_I], df.mean()[1:MAX_I], c='r', alpha=1)
    ax.boxplot(df.iloc[:, 1:MAX_I].T)
    if ylog: ax.set_yscale('log')
    fig.savefig(png, dpi=300, bbox_inches='tight')
```
